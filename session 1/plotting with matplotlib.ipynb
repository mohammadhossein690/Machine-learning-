# Matplotlib :

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
x = np.linspace(0,5,10)
y = x * 2
```

```python
print(x)
print(y)
```

```python
y = x ** 2
```

```python
fig = plt.figure(figsize=(8,3))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig = plt.figure(figsize=(10,3))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='m',linewidth=3,label='y = x ** 2')
axes.set_xlabel('Time',fontsize=15)
axes.set_ylabel('Amplitude',fontsize=15)
axes.set_title('Visualization of y = x ** 2',fontsize=15)
plt.legend(fontsize=15)
plt.show()
```

```python
t = np.linspace(0,10,100)
A = 5
f = 50
sine_wave = A * np.sin(2 * np.pi * f * t)
```

```python
fig = plt.figure(figsize=(16,5))
axes = fig.add_axes([0,0,1,1])
axes.plot(t,sine_wave,linewidth=3,label='sineWave')
axes.set_xlabel('Time',fontsize=20)
axes.set_ylabel('Amplitude',fontsize=20)
axes.set_title('Visualization of Sine Wave',fontsize=25)
axes.set_xlim(0,10)
axes.set_ylim(-5,5)
plt.legend(fontsize=15)
plt.show()
```

```python
fig = plt.figure(figsize=(16,5))
axes = fig.add_axes([0,0,1,1])
axes.plot(t,sine_wave,linewidth=3,label='sineWave')
axes.set_xlabel('Time',fontsize=20)
axes.set_ylabel('Amplitude',fontsize=20)
axes.set_title('Visualization of Sine Wave',fontsize=25)
axes.set_xlim(0,10)
axes.set_ylim(-5,5)
axes.set_xticks([0,10])
axes.set_yticks([-5,0,5])
plt.legend(fontsize=15)
plt.show()
```

```python
fig = plt.figure(figsize=(16,5))
axes = fig.add_axes([0,0,1,1])
axes.plot(t,sine_wave,linewidth=3,label='sineWave')
axes.set_xlabel('Time',fontsize=20)
axes.set_ylabel('Amplitude',fontsize=20)
axes.set_title('Visualization of Sine Wave',fontsize=25)
axes.set_xlim(0,10)
axes.set_ylim(-5,5)
plt.legend(fontsize=15)
plt.axis('off')
plt.show()
```

```python
style.use('dark_background')
fig = plt.figure(figsize=(16,5))
axes = fig.add_axes([0,0,1,1])
axes.plot(t,sine_wave,lw=3,label='sineWave',color='red',ls='--',marker='^',markersize=10,
         markerfacecolor='green',markeredgecolor='yellow')
axes.set_xlabel('Time',fontsize=20)
axes.set_ylabel('Amplitude',fontsize=20)
axes.set_title('Visualization of Sine Wave',fontsize=25)
axes.set_xlim(0,10)
axes.set_ylim(-5,5.5)
plt.legend(fontsize=15)
plt.grid()
plt.show()
```

```python
t = np.arange(0,1,1/300)

x1 = 5 * np.sin(2 * np.pi * 2 * t)

x2 = 2 * np.sin(2 * np.pi * 4 * t)

x3 = 7 * np.sin(2 * np.pi * 6 * t)

x4 = x1 + x2 + x3
```

```python
fig , ax = plt.subplots(nrows=2,ncols=2,figsize=(24,14))
plt.suptitle('Constructing a Wave with different Sine Waves',fontsize=40)
ax[0][0].plot(t,x1,color='red',lw=3)
ax[0][0].set_xlabel('time in sec', fontsize = 20)
ax[0][0].set_ylabel('Amplitude', fontsize = 20)
ax[0][0].set_title("First SineWave", fontsize = 25)
#-------------------------------------------------
ax[0][1].plot(t,x2,color='yellow',lw=3)
ax[0][1].set_xlabel('time in sec', fontsize = 20)
ax[0][1].set_ylabel('Amplitude', fontsize = 20)
ax[0][1].set_title("Second SineWave", fontsize = 25)
#-------------------------------------------------
ax[1][0].plot(t,x3,color='green',lw=3)
ax[1][0].set_xlabel('time in sec', fontsize = 20)
ax[1][0].set_ylabel('Amplitude', fontsize = 20)
ax[1][0].set_title("Third SineWave", fontsize = 25)
#-------------------------------------------------
ax[1][1].plot(t,x4,color='magenta',lw=3)
ax[1][1].set_xlabel('time in sec', fontsize = 20)
ax[1][1].set_ylabel('Amplitude', fontsize = 20)
ax[1][1].set_title("Suumation of other Signals", fontsize = 25)
plt.tight_layout()
plt.show()
```

```python
t = np.arange(0,1,1/300)

x1 = 5 * np.sin(2 * np.pi * 2 * t)

x2 = 2 * np.sin(2 * np.pi * 4 * t)

x3 = 7 * np.sin(2 * np.pi * 6 * t)

x4 = x1 + x2 + x3
```

```python
fig , ax = plt.subplots(nrows=2,ncols=2,figsize=(24,14))
style.use('dark_background')

plt.rcParams['lines.linewidth'] = 3
plt.rcParams['lines.linestyle'] = '--'
plt.rcParams['xtick.labelsize'] = 40
plt.rcParams['ytick.labelsize'] = 40


plt.suptitle('Constructing a Wave with different Sine Waves',fontsize=40)
ax[0][0].plot(t,x1,color='red')
ax[0][0].set_xlabel('time in sec', fontsize = 20)
ax[0][0].set_ylabel('Amplitude', fontsize = 20)
ax[0][0].set_title("First SineWave", fontsize = 25)
#-------------------------------------------------
ax[0][1].plot(t,x2,color='yellow')
ax[0][1].set_xlabel('time in sec', fontsize = 20)
ax[0][1].set_ylabel('Amplitude', fontsize = 20)
ax[0][1].set_title("Second SineWave", fontsize = 25)
#-------------------------------------------------
ax[1][0].plot(t,x3,color='green')
ax[1][0].set_xlabel('time in sec', fontsize = 20)
ax[1][0].set_ylabel('Amplitude', fontsize = 20)
ax[1][0].set_title("Third SineWave", fontsize = 25)
#-------------------------------------------------
ax[1][1].plot(t,x4,color='magenta')
ax[1][1].set_xlabel('time in sec', fontsize = 20)
ax[1][1].set_ylabel('Amplitude', fontsize = 20)
ax[1][1].set_title("Suumation of other Signals", fontsize = 25)
plt.tight_layout()
plt.show()
```
